```python
import rdkit
from rdkit import Chem
from rdkit.Chem import Draw, AllChem
import pandas as pd
from rdkit.Chem.Draw import IPythonConsole
```

```python
# Load checkpoint molecules

st_dir = '../data/starters_targets/'
fns = ['mvacid', '2mg', 'hopa', 'chpd']
dfs = []
for fn in fns:
    p = st_dir + fn + '.csv'
    dfs.append(pd.read_csv(p, delimiter=','))

tdf = pd.concat(dfs, ignore_index=True)
tdf.head()
```

```python
mols = [Chem.MolFromSmiles(smi) for smi in tdf.loc[:, 'smiles']]
labels = [name for name in tdf.loc[:, 'id']]

img = Draw.MolsToGridImage(mols, molsPerRow=1, subImgSize=(200, 120), legends=labels)
with open('../artifacts/checkpoint_mols_hmpa_2mg_chpd.png', 'wb') as f:
    f.write(img.data)

img
```
